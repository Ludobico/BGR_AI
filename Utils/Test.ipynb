```python
DeseaseList = []
for file_name in os.listdir(labeldir):
    file_path = os.path.join(labeldir, file_name)
    with open(file_path, 'r') as f:
        data = json.load(f)

    for i in range(len(data['categories'])):
        DeseaseList.append(data['categories'][i]['name'])
print(DeseaseList)
```

```python
count_dict = {}

for elem in DeseaseList:
    if elem not in count_dict:
        count_dict[elem] = 1
    else:
        count_dict[elem] += 1

for k,v in count_dict.items():
    print(k,v)
```
